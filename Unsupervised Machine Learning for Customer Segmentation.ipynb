## Obtain the optimal number of clusters using the elbow method

```python
scaler=StandardScaler()
```

```python
df_scaled=scaler.fit_transform(df)
```

```python
scores_1=[]
range_values=range(1,20)
```

```python
for i in range_values:
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_scaled)
    scores_1.append(kmeans.inertia_)

plt.plot(scores_1)
```

## Apply k-means using Scikit-Learn to perform customer segmentation

## Apply Principal Component Analysis (PCA) technique to perform dimensionality reduction and data visualization

```python
pca=PCA(n_components=2)
principal_comp=pca.fit_transform(df_scaled)
principal_comp
```

```python
pca_df=pd.DataFrame(data=principal_comp,columns=['pca1','pca2'])
pca_df.head()
```

```python
pca_df=pd.concat([pca_df,pd.DataFrame({'cluster':labels})],axis=1)
pca_df.head()
```

```python
plt.figure(figsize=(10,8))
ax=sns.scatterplot(x='pca1',y='pca2',hue='cluster',data=pca_df,palette=['red','green','blue','yellow','orange','purple','black'])
plt.show()
```
